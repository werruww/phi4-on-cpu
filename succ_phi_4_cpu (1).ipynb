```python
from transformers import pipeline
import torch

messages = [
    {"role": "user", "content": "Who are you?"},
]

pipe = pipeline(
    "text-generation",
    model="microphi-4soft/",
    trust_remote_code=True,
    torch_dtype=torch.bfloat16,  # Changed 'type' to 'torch_dtype'
    device_map="auto"  # Changed 'device_map' to use CPU
)
result = pipe(messages[0]["content"])  # Changed 'pipe(messages)' to 'pipe(messages[0]["content"])'
print(result[0]['generated_text'])
```
